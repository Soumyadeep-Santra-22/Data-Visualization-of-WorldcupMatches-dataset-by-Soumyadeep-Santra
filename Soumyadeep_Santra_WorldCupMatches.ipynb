```python
missing_count = df.isnull().sum()## Count missing values in each column

print(missing_count)
```

```python
# Replace missing values with mean
df['Attendance'] = df['Attendance'].fillna(df['Attendance'].mean())
print(df['Attendance'])

```

```python
## Replace missing values with median
df['Attendance'] = df['Attendance'].fillna(df['Attendance'].median())
print(df['Attendance'])
```

```python
# Verify the changes
print(df['Attendance'].isnull().sum())
```

```python
# Replace missing values in multiple columns with mean
df[['Home Team Goals', 'Away Team Goals', 'Attendance']] = df[['Home Team Goals', 'Away Team Goals', 'Attendance']].apply(lambda x: x.fillna(x.mean()))
print(df[['Home Team Goals', 'Away Team Goals', 'Attendance']])
```
